```python
# use a for loop to cycle throught the review pages
salary_table = []
a = "http://www.espn.com/nba/salaries/_/year/2017/page/"

for i in range(1,13):
    url = '%s%s' %(a,i)
    r = requests.get(url)
    soup = BeautifulSoup(r.content)
    
    g_data = soup.find_all("tr", {"class": "oddrow"})
    for item in g_data:
        table1 = item.text
        salary_table.append(table1)    
        print table1
    g_data1 = soup.find_all("tr", {"class": "evenrow"})
    for item in g_data1:
        table2 = item.text
        salary_table.append(table2)
        print table2
```
