# Data Preparation

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import csv
import numpy as np
```

```python
df = pd.read_csv("data/association.csv")
df.head()
```

The data is in tabular format with zeros and ones. The data needs to be transformed.

```python
df.info()
```

```python
for row in df.values:
    print row
```

```python
for row in df.values:
    print df.columns[row.astype(bool)].tolist()
```

```python
# convert the data and save using a variable

data = [df.columns[row.astype(bool)].tolist() for row in df.values]
data
```

### First, transform data into edgelists

```python
for i in data:
    print i
```

```python
# generating network data

from itertools import combinations

nodes = [x for d in data for x in combinations(d, 2)]
nodes
```

```python
# You could generate a csv file and do network visualization using Gephi

df = pd.DataFrame(nodes)
df.to_csv("data/networkvisualization_2.csv", index=False, header=None)
```
