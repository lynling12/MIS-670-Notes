# Lists
Each value is assigned into an **index** or a **position** ... this is like **address** ... **the first index is zero; the second index is one**

Lists can be thought of as a series of boxes. Each box has a different value. For example, the boxes created by demolist = ['life', 42, 'the universe', 6, 'and', 9] would look like this:

<img src="images\list.gif">

Each box is referenced by its number so the statement demolist[0] would get 'life', demolist[1] would get 42 and so on up to demolist[5] getting 9.

```python
l = [ "Drake", "Derp", "Derek", "Dominique", 42, 6, "Universe", 9]
 
print l     # prints all elements
print l[0]  # print first element ... each value is assigned into an index or a position ... the first index is zero
print l[1]  # prints second element ... the second index is one
print l[0:2] # print the first two items
print l[-1]  # print the last item
```

```python
len(l)
```
