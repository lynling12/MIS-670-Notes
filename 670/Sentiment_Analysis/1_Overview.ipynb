# Types of Sentiment Analysis
#### Supervised Sentiment Analysis
- Basically sames as classification analysis (or text classification) where labels are: 
    - positive and negative
    - fresh and rotten (http://www.rottentomatoes.com/m/star_wars_episode_vii_the_force_awakens/)                        

#### Unsupervised Sentiment Analysis
- No label information in the dataset (e.g., tweets, product reviews, resturant reviews)
- Use **lexicons** (list of positive and negative words)

#### Lexicons
- Bing Liu Lexicon (2_Ipython Notebook)
    - Lists of positive words and negative words
    - Per each customer review, count the number of positive and negative words appearing
- Pattern (3_Ipython Notebook)
    - Python package for sentiment analysis
    - Including a lexicon assigning positive and negative scores to words
- SentiStrength (4_Ipython Notebook)
    - http://sentistrength.wlv.ac.uk/ 
    - “SentiStrength estimates the strength of positive and negative sentiment in short texts, even for informal language. It has human-level accuracy for short social web texts in English, except political texts. SentiStrength reports two sentiment strengths:
    - -1 (not negative) to -5 (extremely negative); 1 (not positive) to 5 (extremely positive)”
